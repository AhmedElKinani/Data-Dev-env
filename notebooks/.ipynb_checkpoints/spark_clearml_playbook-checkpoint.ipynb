## 1. Imports, Environment Setup & Configurations

First, we'll read environment variables (configured automatically via docker-compose) to setup our PostgreSQL connection strings and locate Spark.

````python
import os
import sys
import pandas as pd

# 1. PostgreSQL connection environment variables
PG_USER     = os.environ.get('POSTGRES_USER',     'postgres')
PG_PASSWORD = os.environ.get('POSTGRES_PASSWORD', 'postgres')
PG_HOST     = os.environ.get('POSTGRES_HOST',     'postgres')
PG_PORT     = os.environ.get('POSTGRES_PORT',     '5432')
PG_DB       = os.environ.get('POSTGRES_DB',       'postgres')

# JDBC configurations
JDBC_URL   = f'jdbc:postgresql://{PG_HOST}:{PG_PORT}/{PG_DB}'
JDBC_PROPS = {
    'user': PG_USER, 
    'password': PG_PASSWORD, 
    'driver': 'org.postgresql.Driver'
}
JDBC_JAR = os.environ.get('SPARK_JDBC_JAR', '/home/jovyan/jars/postgresql-42.7.3.jar')

print("Environment Configurations Loaded:")
print(f"  JDBC URL    : {JDBC_URL}")
print(f"  JDBC Driver : {JDBC_JAR}")
````

## 2. ClearML Task Tracking & Session Initialization

We initialize a ClearML task to track this run. The credentials in Jupyter's environment variables (`CLEARML_API_ACCESS_KEY` & `CLEARML_API_SECRET_KEY`) allow Jupyter to connect immediately without manual setup.

````python
from clearml import Task

# Initialize task on the local ClearML server
task = Task.init(
    project_name='DataStackPlaybook',
    task_name='ultimate_cheatsheet_run',
    task_type=Task.TaskTypes.training,
    reuse_last_task_id=False,
)

# Log configuration settings to ClearML
run_config = {
    'database_name': PG_DB,
    'jdbc_driver': JDBC_JAR,
    'spark_master': 'local[*]',
    'model_type': 'LinearRegression',
    'random_seed': 42
}
task.connect(run_config)

print(f"ClearML Task successfully initialized. Task ID: {task.id}")
print(f"Inspect metrics at: http://localhost:8080/projects/ (ClearML Web UI)")
````

## 3. ClearML ML Model Governance (Official API Reference)

Model governance ensures that every model's lineage, configuration, files, and deployment stages (Draft, Testing, Production, Archive) are tracked programmatically. Below are the exact code snippets required for each type of connection.

---

### A. Automatic Logging (Implicit Connection)
ClearML automatically binds to many frameworks (e.g. Scikit-Learn, PyTorch, TensorFlow, XGBoost). Once `Task.init()` is called, saving a model using the framework's native API will automatically register and upload it to ClearML.

```python

```

### B. Manual Model Registration (`OutputModel`)
Use `OutputModel` to manually register a model, link it to the active task, upload weights, specify labels, and set custom metadata.

````python
from clearml import OutputModel

# 1. Initialize output model and link to active task
output_model = OutputModel(task=task, name="airfoil_noise_prediction_model", framework="PySpark")

# 2. Update model files/weights (Point to local path or directory containing saved model files)
# Note: Call this after saving the Spark model files in Section 7
# output_model.update_weights(weights_filename="Final_Project")

# 3. Update model labels (useful for classification models to map class indices to names)
output_model.update_labels({"Frequency": 0, "AngleOfAttack": 1})

# 4. Update design/evaluation metrics
output_model.update_design_metrics({
    "r2": 0.5426,
    "mse": 22.5937,
    "mae": 3.7336
})

# 5. Set custom metadata parameters
output_model.set_metadata({
    "author": "Antigravity Data Engineer",
    "dataset_source": "NASA Airfoil Noise Dataset",
    "pipeline_stages_count": 3
})

print(f"Manual OutputModel initialized. Model ID: {output_model.id}")
````

### C. Reusing Registered Models for Lineage (`InputModel`)
To load a previously registered model (for evaluation, fine-tuning, or inference) and preserve lineage.

````python
from clearml import InputModel

# 1. Connect an input model by its ID to the current task to preserve lineage
input_model = InputModel(model_id=output_model.id)
task.connect(input_model)

# 2. Download and cache model files locally
local_weights_path = input_model.get_weights()
print(f"Downloaded input model weights cached locally at: {local_weights_path}")
````

### D. Model Registry Queries & Status Transitions (Governance States)

````python
from clearml import Model

# 1. Query models from the registry based on project, tags, or status
models = Model.query_models(
    project_name="DataStackPlaybook",
    tags=["staging"],
    only_published=False
)
print(f"Found {len(models)} models matching query.")

# 2. Fetch a model by ID and update tags / description programmatically
reg_model = Model(model_id=output_model.id)
reg_model.tags = ["staging", "v1.0"]

# Add comments
reg_model.comment = "Final airfoil prediction model build"

# Publish the model (sets status to Published, making it read-only for strict governance)
reg_model.publish()

# Alternatively: Archive model if deprecated
# reg_model.archive()

print(f"Model {reg_model.name} (ID: {reg_model.id}) status:")
print(f"  Published: {reg_model.published}")
print(f"  Tags:      {reg_model.tags}")
print(f"  Comment:   {reg_model.comment}")
````

## 4. Start PySpark Session

We search for Spark on the system path using `findspark`, then configure and start the `SparkSession` loading the Postgres JDBC driver jar.

````python
import findspark
findspark.init()

from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.types import StructType, StructField, StringType, DoubleType, IntegerType

spark = (
    SparkSession.builder
    .appName('DataStackPlaybook-Spark')
    .master('local[*]')
    .config('spark.jars', JDBC_JAR)
    .config('spark.sql.shuffle.partitions', '8')
    .getOrCreate()
)
spark.sparkContext.setLogLevel('WARN')

print(f"Spark {spark.version} Session created successfully!")
````

## 5. Spark ETL & DataFrame API Reference (Based on Final_Project.ipynb)

This section shows programmatic ETL tasks including loading CSVs with options, filtering, row counts, dropping duplicates, removing null values, column renaming, schema type casting, and writing/reading Parquet format.

````python
import urllib.request

# 1. Download the raw CSV airfoil dataset from IBM IBM-S3 Storage
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMSkillsNetwork-BD0231EN-Coursera/datasets/NASA_airfoil_noise_raw.csv"
csv_path = "NASA_airfoil_noise_raw.csv"
print(f"Downloading dataset from {url}...")
urllib.request.urlretrieve(url, csv_path)
print("Download completed.")

# 2. Load CSV dataset with schema inference and headers
df_raw = spark.read.csv(csv_path, header=True, inferSchema=True)

# 3. Print Row count and Schema
rowcount_raw = df_raw.count()
print(f"Total raw rows loaded: {rowcount_raw}")
df_raw.printSchema()

# 4. Display top 5 rows
print("Top 5 raw rows:")
df_raw.show(5)

# 5. Drop Duplicates
df_no_dupes = df_raw.dropDuplicates()  # can also use drop_duplicates()
rowcount_no_dupes = df_no_dupes.count()
print(f"Rows after dropping duplicates: {rowcount_no_dupes}")

# 6. Drop rows with null values
df_clean = df_no_dupes.dropna()
rowcount_clean = df_clean.count()
print(f"Rows after dropping nulls: {rowcount_clean}")

# 7. Rename Columns (e.g. rename target column 'SoundLevel' to 'SoundLevelDecibels')
df_renamed = df_clean.withColumnRenamed("SoundLevel", "SoundLevelDecibels")

# 8. Type Casting / Column manipulation
# E.g., casting 'Frequency' column to double for math compatibility
df_transformed = df_renamed.withColumn("FrequencyDouble", F.col("Frequency").cast("double"))

# 9. Save the Cleaned DataFrame in Parquet Format
parquet_path = "NASA_airfoil_noise_cleaned.parquet"
print(f"Saving cleaned dataset as Parquet to {parquet_path}...")
df_transformed.write.mode("overwrite").parquet(parquet_path)
print("Parquet save completed.")

# 10. Load the Parquet file back to confirm
df = spark.read.parquet(parquet_path)
print(f"Successfully loaded {df.count()} rows back from Parquet file.")
````

## 6. In-Database Manipulation using SparkSQL

We can register our loaded dataset as a temporary view inside Spark and run standard SQL queries to transform and filter data.

````python
# 1. Register as a temporary SQL view
df.createOrReplaceTempView("v_nasa_airfoil")

# 2. Run SparkSQL query to discover high noise events (> 125 dB) for frequencies >= 1000
sql_df = spark.sql("""
    SELECT 
        Frequency,
        AngleOfAttack,
        ChordLength,
        FreeStreamVelocity,
        SoundLevelDecibels,
        CASE 
            WHEN SoundLevelDecibels > 130 THEN 'Extreme'
            WHEN SoundLevelDecibels > 120 THEN 'Moderate'
            ELSE 'Low'
        END as NoiseClass
    FROM v_nasa_airfoil
    WHERE Frequency >= 1000 AND SoundLevelDecibels >= 125.0
    ORDER BY SoundLevelDecibels DESC
""")

print(f"Discovered {sql_df.count()} high noise records using SparkSQL:")
sql_df.show(5)
````

## 7. PySpark Machine Learning (MLlib) Pipeline (Airfoil Noise Prediction)

This section builds a complete machine learning pipeline using Spark MLlib. It demonstrates feature engineering (`VectorAssembler`, `StandardScaler`), regression modeling (`LinearRegression`), model training/evaluation, coefficient extraction, and model persistence.

````python
from pyspark.ml import Pipeline
from pyspark.ml.pipeline import PipelineModel
from pyspark.ml.feature import VectorAssembler, StandardScaler, StringIndexer, OneHotEncoder
from pyspark.ml.regression import LinearRegression
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import RegressionEvaluator, MulticlassClassificationEvaluator

# --- A. FEATURE ENGINEERING STAGES ---
# 1. VectorAssembler: Combine features into a single vector column
feature_cols = ['Frequency', 'AngleOfAttack', 'ChordLength', 'FreeStreamVelocity', 'SuctionSideDisplacement']
assembler = VectorAssembler(
    inputCols=feature_cols,
    outputCol="features"
)

# 2. StandardScaler: Scale feature values to have unit standard deviation
scaler = StandardScaler(
    inputCol="features", 
    outputCol="scaledFeatures",
    withStd=True, 
    withMean=False
)

# Reference for Categorical Features (uncomment if working with categorical string variables):
# indexer = StringIndexer(inputCol="categorical_string_col", outputCol="indexed_col")
# encoder = OneHotEncoder(inputCol="indexed_col", outputCol="ohe_vector_col")

# --- B. MACHINE LEARNING MODEL STAGE ---
# Define LinearRegression stage (scaledFeatures as input, SoundLevelDecibels as target label)
lr = LinearRegression(
    featuresCol="scaledFeatures", 
    labelCol="SoundLevelDecibels"
)

# --- C. PIPELINE ASSEMBLY ---
# Build pipeline containing assembler, scaler, and estimator
pipeline = Pipeline(stages=[assembler, scaler, lr])

# --- D. TRAIN/TEST SPLIT ---
# Split the data (70% training, 30% testing) using seed=42
(trainingData, testingData) = df.randomSplit([0.7, 0.3], seed=42)
print(f"Training set rows: {trainingData.count()}, Testing set rows: {testingData.count()}")

# --- E. FIT PIPELINE MODEL ---
pipelineModel = pipeline.fit(trainingData)
print("ML Pipeline successfully fitted on training data.")

# --- F. INFERENCE / PREDICTION ---
# Run prediction transform on testingData
predictions = pipelineModel.transform(testingData)
print("Sample Predictions Output:")
predictions.select("SoundLevelDecibels", "prediction").show(5)

# --- G. MODEL EVALUATION ---
# Instantiate regression evaluators for MSE, MAE, and R-Squared
eval_mse = RegressionEvaluator(predictionCol="prediction", labelCol="SoundLevelDecibels", metricName="mse")
eval_mae = RegressionEvaluator(predictionCol="prediction", labelCol="SoundLevelDecibels", metricName="mae")
eval_r2  = RegressionEvaluator(predictionCol="prediction", labelCol="SoundLevelDecibels", metricName="r2")

mse = eval_mse.evaluate(predictions)
mae = eval_mae.evaluate(predictions)
r2  = eval_r2.evaluate(predictions)

print("\n--- Evaluation Metrics ---")
print(f"Mean Squared Error (MSE) : {mse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"R-Squared (R2)           : {r2:.4f}")

# --- H. EXTRACT STAGE COEFFICIENTS ---
lrModel = pipelineModel.stages[-1]
print(f"\nModel Intercept: {lrModel.intercept:.4f}")
print("Model Coefficients:")
for col_name, coef in zip(feature_cols, lrModel.coefficients):
    print(f"  {col_name}: {coef:.4f}")

# --- I. MODEL PERSISTENCE ---
model_save_path = "Final_Project_Model"
print(f"\nSaving PySpark PipelineModel to disk: {model_save_path}...")
pipelineModel.write().overwrite().save(model_save_path)
print("Save complete.")

# Load it back to verify correctness
loadedPipelineModel = PipelineModel.load(model_save_path)
loaded_predictions = loadedPipelineModel.transform(testingData)
print(f"Loaded model verified. Row count of loaded predictions: {loaded_predictions.count()}")
````

## 8. PostgreSQL Database Operations (JDBC)

Now we write our regression predictions table into PostgreSQL and verify the write by loading it back into a new Spark DataFrame.

````python
# 1. Write the prediction metrics to PostgreSQL via JDBC
output_table_name = "airfoil_noise_predictions"
print(f"Writing data to database table: {output_table_name}...")

predictions.select("SoundLevelDecibels", "prediction").write.jdbc(
    url=JDBC_URL,
    table=output_table_name,
    mode="overwrite",
    properties=JDBC_PROPS
)
print("Write completed successfully!")

# 2. Read the table back from PostgreSQL to confirm
df_loaded = spark.read.jdbc(
    url=JDBC_URL,
    table=output_table_name,
    properties=JDBC_PROPS
)
print(f"Successfully loaded {df_loaded.count()} rows back from the database table '{output_table_name}':")
df_loaded.show(5)
````

## 9. Final Log & Register Model to ClearML

Finally, we connect our final trained models and metrics to ClearML Model Registry for governance, and perform cleanup.

````python
# 1. Report evaluation metrics to ClearML task logger
logger = task.get_logger()
logger.report_single_value(name="Evaluation MSE", value=mse)
logger.report_single_value(name="Evaluation MAE", value=mae)
logger.report_single_value(name="Evaluation R2", value=r2)
logger.report_single_value(name="Intercept", value=lrModel.intercept)

# 2. Connect and upload the saved model directory to the ClearML Registry
print("Registering model weights with ClearML OutputModel...")
output_model.update_weights(weights_filename="Final_Project_Model")

# Update design metrics matching the actual run
output_model.update_design_metrics({
    "r2": r2,
    "mse": mse,
    "mae": mae
})

# Publish model to seal it as ready-to-use and lock its metadata
output_model.publish()

print(f"Governance successfully completed. Model registered under ID: {output_model.id}")
````

## 10. Cleanup & Teardown

Finally, we'll terminate the Spark session and close our ClearML task to complete the pipeline.

````python
# Stop Spark Session
spark.stop()
print("Spark session stopped.")

# Close ClearML Task
task.close()
print("ClearML task closed.")
````
